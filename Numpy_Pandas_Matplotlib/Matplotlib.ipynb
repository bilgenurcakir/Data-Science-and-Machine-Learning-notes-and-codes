```python
np_array1=np.linspace(0,10,20) # 0dan 10 a kadar 20 sayı don araları eşit olsun
np_array1
```

```python
np_array2=np_array1 ** 3
np_array2
```

```python
plt.plot(np_array1,np_array2,"b*-")
plt.show()
```

```python
#birden fazla grafiği yan yana görme
plt.subplot(1,2,1)
plt.plot(np_array1,np_array2,"b*-")
plt.subplot(1,2,2)
plt.plot(np_array2,np_array1,"g*-")
plt.show()
```

```python
my_figure=plt.figure()
figure_axes=my_figure.add_axes([0.1,0.1,0.7,0.7])
figure_axes.plot(np_array1,np_array2,"g")
figure_axes.set_xlabel("x axis")
figure_axes.set_ylabel("y axis")
plt.show()
```

```python
my_figure=plt.figure(dpi=100) #çözünürlük
figure_axes=my_figure.add_axes([0.1,0.1,0.9,0.9])
figure_axes.plot(np_array1 ,np_array1 ** 2 ,"g",label="iki üssü")
figure_axes.plot(np_array1 ,np_array1 ** 3 ,"b" ,label="3 üssü")
figure_axes.legend()
figure_axes.set_xlabel("x axis")
figure_axes.set_ylabel("y axis")
plt.show()
```

```python
data1=np.linspace(0,10,20)
data2=data1 **2
```

```python
fig,axes=plt.subplots() #figure ve eksen
```

```python
fig,axes=plt.subplots()
axes.plot(data1,data2,color="#1E0703" ,alpha=0.5)
axes.plot(data2,data1,color="#53CA0A")
plt.show()
```

```python
new_fig,new_axes=plt.subplots()
new_axes.plot(data1,data1+2,color="b", linewidth=1.2)
new_axes.plot(data1,data1+3, color="y", linewidth=1.3)
new_axes.plot(data1,data1+5, color="g", linestyle="--")
new_axes.plot(data1,data1+7, color="blue", linestyle="--", marker="o", markersize=4,markerfacecolor="r")
plt.show()
```

```python
#scatter

plt.scatter(data1,data2)
plt.show()
```
