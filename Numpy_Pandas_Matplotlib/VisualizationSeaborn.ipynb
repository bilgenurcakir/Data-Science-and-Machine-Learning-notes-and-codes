```python
#Seaborn
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x="Height",y="Weight",data=data)
plt.xlabel=("height of athlete")
plt.ylabel=("weight of athlete")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x="Height",y="Weight", hue="Sex",data=data)  
#cinsiyeti farkl renklerle gster
plt.xlabel=("height of athlete")
plt.ylabel=("weight of athlete")
plt.show()   
```

```python
data.columns
```

```python
data["Medal"].unique() # o kolondaki deerleri gsterir
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x="Height",y="Weight", hue="Sex", style="Medal",data=data) 

#cinsiyeti farkl renklerle gster, ald madalyay farkl ekilde gster (* o vb)
plt.xlabel=("height of athlete")
plt.ylabel=("weight of athlete")
plt.show()   
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x="Height",y="Weight", hue="Sex", style="Medal", size="Age",data=data) 

#cinsiyeti farkl renklerle gster, ald madalyay farkl ekilde gster (* o vb)
# noktalar bydke ya buyuyor
plt.xlabel=("height of athlete")
plt.ylabel=("weight of athlete")
plt.show()   
```

```python
sns.set_style("whitegrid")
sns.lineplot(x="Height",y="Weight", hue="Sex", data=data)

plt.xlabel=("height of athlete")
plt.ylabel=("weight of athlete")
plt.show()   
```

```python
sns.set_style("white")
sns.displot(x="Height", hue="Sex", data=data)
plt.ylabel=("frequency")
plt.title("athletes height distribution")
```

```python
sns.set_style("white")
sns.displot(x="Height", hue="Sex", data=data, kind="kde")
plt.ylabel=("frequency")
plt.title("athletes height distribution")
```

```python
sns.barplot(x="Medal" ,y="Height", data=data, hue="Sex")
plt.show()
```

```python
sns.catplot(x="Medal" ,y="Height", data=data, hue="Sex",col="Season")
plt.show()
```
