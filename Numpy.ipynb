```python
# z-score
```

```python
data=np.array([10,12,13,15,18,25,100,105])
#outlier: aykırı değerler 100-105
```

```python
mean=np.mean(data)
mean
```

```python
std=np.std(data)
std
```

```python
z_score= (data- mean)/std
z_score
```

```python
#outlierların 1 den büyük olduğu yerler outlierlar
```

```python
data[z_score <1]
```

```python
#math equations
# 2x + 3y = 8 and 5x + 7y = 19

#coefficient : katsayılar
A=np.array([[2,3],[5,7]])

#constant: sabitler
b=np.array([8,19])

solution=np.linalg.solve(A,b)
solution
```
