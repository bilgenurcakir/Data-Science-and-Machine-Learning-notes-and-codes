```python
#merge
```

```python
dfmerge1=pd.read_csv("7-merge_data1.csv")
dfmerge1
```

```python
dfmerge2=pd.read_csv("7-merge_data2.csv")
dfmerge2
```

```python
#kesisen sutunlar var, farklı sutunlar var
```

```python
#outer join
```

```python
df_outer=pd.merge(dfmerge1,dfmerge2,on="Employee_ID", how="outer") #on: kesişen sutun
```

```python
df_outer #bulamadıklarına NaN atadı
```

```python
# left join
```

```python
df_left=pd.merge(dfmerge1,dfmerge2,on="Employee_ID",how="left")
```

```python
df_left # ilk verilen data frame'e göre yaz. ikinci df'de ona uygun kısımları yaz, diğerlerini yazma 
```

```python
#right join
```

```python
df_right=pd.merge(dfmerge1,dfmerge2,on="Employee_ID", how="right")
```

```python
df_right # ikinci verilen data frame'e göre yaz. ilk df'de ona uygun kısımları yaz, diğerlerini yazma 
```

```python
#inner join
```

```python
df_inner=pd.merge(dfmerge1,dfmerge2,on="Employee_ID", how="inner")
```

```python
df_inner # her iki tablodada eşleşen, dolu olanları al kalanları bırak
```

```python
def Salary_category(salary):
    if salary< 50000:
        return "low"
    elif 50000<= salary < 80000:
        return "medium"
    else:
        return "high"
```

```python
df["Salary"].apply(Salary_category)
```

```python
df["Salary Category"]=df["Salary"].apply(Salary_category)
```

```python
df
```

```python
def new_performance_score(row):
    if row["Experience"] > 10 :
        return  row["Performance_Score"]+1
    else:
        return row["Performance_Score"]
```

```python
df["adjusted"]=df.apply(new_performance_score,axis=1)
df
```

```python
df["formatted name"]=df["Name"].apply(lambda x: x.replace("_"," "))
df
```
